**Load joblib version**

```python
import pandas as pd
from joblib import load
```

```python
joblib_path = "D:\Machine learning Udemy\ML model deployment_9\Saving and loading_2\model_dir\diabetes_model_pipeline.joblib"
```

```python
# load the saved model
loaded_joblib_pipeline = load(joblib_path)
```

```python
# example new patient data
new_data = pd.DataFrame({
    "Pregnancies": [2],
    "Glucose": [120],
    "BloodPressure": [70],
    "SkinThickness": [25],
    "Insulin": [80],
    "BMI": [28.5],
    "DiabetesPedigreeFunction": [0.45],
    "Age": [35]
})

display(new_data)
```

```python
# data will be scaled and passed on to the model for prediction inside the pipeline
prediction = loaded_joblib_pipeline.predict(new_data)
print("Predicted class:", prediction[0])

if prediction[0] == 1:
    print("⚠️ Model predicts: Diabetes")
else:
    print("✅ Model predictes: No Diabetes")
```

**Load pickle version**

```python
import pandas as pd
import pickle

pickle_path = "D:\Machine learning Udemy\ML model deployment_9\Saving and loading_2\model_dir\diabetes_model_pipeline_pickle.pkl"

# load the trained pipeline using pickle
with open(pickle_path, "rb") as file:
    loaded_pickle_pipeline = pickle.load(file)

# example new patient data
new_data = pd.DataFrame({
    "Pregnancies": [2],
    "Glucose": [120],
    "BloodPressure": [70],
    "SkinThickness": [25],
    "Insulin": [80],
    "BMI": [28.5],
    "DiabetesPedigreeFunction": [0.45],
    "Age": [35]
})

display(new_data)

# data will be scaled and passed to the model automatically inside the pipeline
prediction = loaded_pickle_pipeline.predict(new_data)
print("Predicted class:", prediction[0])

if prediction[0] == 1:
    print("⚠️ Model predicts: Diabetes")
else:
    print("✅ Model predicts: No Diabetes")
```

| Aspect | joblib | pickle |
|------|--------|--------|
| Speed (large models) | Fast | Slow |
| NumPy array handling | Optimized | Not optimized |
| Memory usage | Efficient | High |
| ML pipelines | Ideal | Works but suboptimal |
| Built-in | No | Yes |
| scikit-learn recommended | Yes | No |
